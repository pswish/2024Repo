```python
import numpy as np 
import cv2 
import matplotlib.pyplot as plt 
from skimage import filters 
```

```python
# Load the image 
image = cv2.imread('/home/pswish/Downloads/wine.jpeg') 
```

```python
# Convert to grayscale 
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) 
```

```python
# Apply Gaussian Blur 
blurred_image = cv2.GaussianBlur(gray_image, (5, 5), 0) 
```

```python
# Display the result 
plt.imshow(blurred_image, cmap='gray') 
plt.axis('off') 
plt.show() 
```

```python
# Apply Median Filter 
median_filtered_image = cv2.medianBlur(gray_image, 5) 
```

```python
# Display the result 
plt.imshow(median_filtered_image, cmap='gray') 
plt.axis('off') 
plt.show() 
```

```python
# Define a threshold value 
threshold_value = 150 

# Apply global thresholding 
_, binary_global = cv2.threshold(gray_image, threshold_value, 255, cv2.THRESH_BINARY) 
```

```python
# Display the result 
plt.imshow(binary_global, cmap='gray') 
plt.axis('off') 
plt.show()
```
